```python
# Import statements
import pandas as pd; import numpy as np; import matplotlib.pyplot as plt
```

```python
# Import data
df = pd.read_csv('nonPolygonData.csv')
```

```python
# Remove data points with null visit counts
df2 = df[df.sg_wp__raw_visit_counts.notna()]
```

```python
# Find the median visit count for each date
plot = df2[['date_range_start','sg_wp__raw_visit_counts']].groupby('date_range_start').agg(np.median).reset_index()
```

```python
# Plot the median visit count vs. date for 2018 to 2022
plot.plot('date_range_start', 'sg_wp__raw_visit_counts')
```
